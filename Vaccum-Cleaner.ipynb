```python
# For 2 Rooms:
goalState = {'A': '0', 'B': '0'}
roomState = {'A': '0', 'B': '0'}
cost = 0

print("Enter the Starting location of BOT {A | B}")
location = input()
print()

for room in roomState:
    roomState[room] = input("Enter the State of " + room + " the room {0: Clean ; 1: Dirty}\n")

while roomState != goalState:
    if location == 'A':  # Bot initially in room A
        if roomState['A'] == '1':  # A is dirty
            roomState['A'] = '0'
            cost += 1
            print("A was Dirty and now is Cleaned\n")

        print("A -> B\n")
        cost += 1
        location = 'B'  # Move to room B

    elif location == 'B':  # Bot initially in room B
        if roomState['B'] == '1':  # B is dirty
            roomState['B'] = '0'
            cost += 1
            print("B was Dirty and now is Cleaned\n")

        print("B -> A\n")
        cost += 1
        location = 'A'  # Move to room A

    print("Current Room State:", roomState)
    print("Cost:", cost)
    print()

print("Goal State has been Achieved\n")
print("Performance Measure: " + str(cost))




```

```python
# For 4 Rooms
goalState = {'A': '0', 'B': '0', 'C': '0', 'D': '0'}
roomState = {'A': '0', 'B': '0', 'C': '0', 'D': '0'}
cost = 0

print("Enter the Starting location of BOT {A | B | C | D}")
location = input()
print()

for room in roomState:
    roomState[room] = input("Enter the State of " + room + " the room {0: Clean ; 1: Dirty}\n")

while roomState != goalState:
    if location in roomState:
        current_room = location

        if roomState[current_room] == '1':
            roomState[current_room] = '0'
            cost += 1
            print(current_room + " was Dirty and now is Cleaned\n")

        next_room = input(current_room + " -> ? Enter the next room {A | B | C | D}\n")

        if next_room in roomState and next_room != current_room:
            print(current_room + " -> " + next_room + "\n")
            cost += 1
            location = next_room

    print("Current Room State:", roomState)
    print("Cost:", cost)
    print()

print("Goal State has been Achieved\n")
print("Performance Measure:", cost)
```
